```python
import matplotlib.pyplot as plt
```

```python
def func(x):
    return (x+3) ** 2

def grad(x):
    return 2*(x+3)
```

```python
x = 2
learning_rate = 0.1
epochs = 50

x_vals = [x]
y_vals = [func(x)]
```

```python
for i in range(epochs):
    dx = grad(x)
    x = x - learning_rate * dx

    x_vals.append(x)
    y_vals.append(func(x))

    print(f"Iteration {i+1}: x = {x:.5f}, f(x) = {func(x):.5f}")
```

```python
print(f"\nLocal minimum at x = {x:.5f}, f(x) = {func(x):.5f}")
```
